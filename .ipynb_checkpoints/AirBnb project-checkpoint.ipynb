# Modelling


The nature of my data set and the target variable `country_destination` can be targeted as a multiclass classification problem. However, due to the shortage of time and to the complexity of the problem, I've decided to consider the problem as a binary classification. The scope of the modelling is to predict wheather a new user will go in the `US` or not. I choose the `US` as a target varible becouse from my EDA I found out that the `US` was the most choosen country by the existing users. 

```python
data_fin["US"] = (data_fin["country_destination"] == "US").astype(int)
```

```python
(data_fin["US"].value_counts(normalize=True) * 100).round().astype(int)# 70% of existing users are choosing US 
```

From the result above I can clearly see that there is class imbalance issue. I will solve the problem 
by applying Synthetic Minority Oversampling(SMOTE). By using SMOTE technique the algorithm will generate new sample data by creating 'synthetic' examples that are the combinations of the closest minority class cases.

```python
X = data_fin.drop(
    ["country_destination",'date_account_created','date_first_active','timestamp_first_active','id','US'],
    axis=1) 
y = data_fin["US"] # my target variable 
```

```python
# Create train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
# Create validation set
X_train_v, X_val, y_train_v, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=1)
```

```python
# Previous original class distribution
print('Original class distribution: \n')
print(y_train.value_counts())
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_sample(X_train, y_train) 
# Preview synthetic sample class distribution
print('-----------------------------------------')
print('Synthetic sample class distribution: \n')
print(pd.Series(y_train_resampled).value_counts()) 
```

```python
# Previous original class distribution
print('Original class distribution: \n')
print(y_train_v.value_counts())
smote = SMOTE()
X_train_v_resampled, y_train_v_resampled = smote.fit_sample(X_train_v, y_train_v) 
# Preview synthetic sample class distribution
print('-----------------------------------------')
print('Synthetic sample class distribution: \n')
print(pd.Series(y_train_v_resampled).value_counts())
```

By using the SMOTE technique above is possible to see that now in the training set the target variable is equally divided in two classes of 17005 observations each. In the validation set the target variable is equally divided in two classes of 14485 observations each.

```python
print(len(X_train_resampled))
print(len(X_test))
print(len(X_val))
```

I will use the funcion below for simplify my modelling 

```python
def roc_plot(model, X_train, y_train, X_val, y_val):
    """"
    This function allows us to plot our roc curve based off our predicted probabilities from our Train and validation sets
    """
    train_prob = model.predict_proba(X_train)[:, 1]
    val_prob = model.predict_proba(X_val)[:, 1]
    plt.figure(figsize=(7, 7))
    for data in [[y_train, train_prob], [y_val, val_prob]]:  # ,[y_test, test_prob]
        fpr, tpr, threshold = roc_curve(data[0], data[1])
        plt.plot(fpr, tpr)
    annot(fpr, tpr, threshold)
    plt.plot([0, 1], [0, 1], color="navy", linestyle="--")
    plt.ylabel("TPR (power)")
    plt.xlabel("FPR (alpha)")
    plt.legend(["train", "test"])
    plt.show()
```

## Logistic regression

`Due to the size of my dataset, I decided to use just Cross Validation on the whole training dataset.`

### Baseline Model

```python
baseline_model = LogisticRegression(fit_intercept=True, random_state=42, max_iter=1000, solver='sag')
```

```python
skf = StratifiedKFold(n_splits = 3, random_state = 600, shuffle=True)
```

```python
cv_result_base_lr = cross_val_score(baseline_model, X_train_resampled, y_train_resampled, 
                                    cv=skf, scoring='roc_auc')
print("Cross Validation Score", round(np.mean(cv_result_base_lr), 3)) 
```

The score [0.772] of the Cross Validation for the baseline Logistic Regression model is not bad but it can be improved.

### Hyperparameter tuning

```python
params_grid = {'solver':['liblinear','sag','saga']}
# {'penalty':['l1','l2']}
# 'C':[1,10,100],
```

```python
opt_model_lr = GridSearchCV(LogisticRegression(random_state=42, max_iter=1000),
                         params_grid,
                         cv=skf,
                         scoring='roc_auc',
                         return_train_score=True,
                         verbose=2,
                         n_jobs=-1)

opt_model_lr.fit(X_train_resampled, y_train_resampled)
```

```python
best_logreg = opt_model_lr.best_estimator_
```

```python
print('best params', opt_model_lr.best_params_)
print()
print('best estimator', opt_model_lr.best_estimator_)
print()
pd.DataFrame(opt_model_lr.cv_results_)
```

```python
cv_result_opt_lr = round(opt_model_lr.best_score_, 3)
print('Best Cross Validation Score', cv_result_opt_lr) # roc_auc cross validation score
```

By applying hyperparameter tuning I can see that the Logistic Regression model is increasing the performance giving a result of 0.804.
